```python
import pandas as pd
df = pd.read_csv('covid_data_22-09-2021_2.csv')
print(df.shape)
```

```python
df = df.drop(['Country'], axis=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
confirmed = df['Confirmed'].values
prop_infected = df['proportion infected'].values
deaths = df['Deaths'].values
gdp = df['GDP'].values
pop_density = df['Population Density'].values
pop = df['2021 Population'].values
area = df['Area'].values
literacy = df['Literacy'].values
employment = df['Employment'].values
old = df['Population 65 +'].values
foreign = df['Foreigners'].values
```

```python
#histogram confirmed
plt.hist(confirmed, bins = 20)
plt.title('Histogram of Confirmed')
plt.ylabel('counts')
plt.xlabel('Confirmed')
plt.show()
```

```python
#histogram proportion infected
plt.hist(prop_infected, bins =20)
plt.title('Histogram of Proportion infected')
plt.ylabel('counts')
plt.xlabel('Proportion infected')
plt.show()
```
